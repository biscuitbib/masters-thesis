```python
import os
import nibabel as nib
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from sklearn.decomposition import PCA
from scipy import ndimage

from thesisproject.utils import create_animation

plt.rcParams['figure.figsize'] = [20, 10]
```

```python
def create_animation(image, dim=0):
    h, w, d = image.shape

    fig, ax = plt.subplots()

    frames = []
    if dim == 0:
        for i in range(h):
            im = ax.imshow(image[i, :, :])
            frames.append([im])
    elif dim == 1:
        for i in range(w):
            im = ax.imshow(image[:, i, :])
            frames.append([im])
    elif dim == 2:
        for i in range(d):
            im = ax.imshow(image[:, :, i])
            frames.append([im])


    ani = animation.ArtistAnimation(fig, frames, interval=50, blit=True, repeat_delay=1000)

    ani.save(f"dim_{dim}.mp4")
```

```python
scan = nib.load("9004175-Right-V05.nii.gz").get_fdata()
scan_flip = np.flip(scan, axis=0)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(scan[:, :, 110])
ax2.imshow(scan_flip[:, :, 110])
```
